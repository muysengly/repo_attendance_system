```python
import win32api
import win32con
import win32gui
```

```python

# Path to your python.exe in venv
exe_path = r"../venv/Scripts/python.exe"
icon_path = r"../asset/my_logo.ico"

# Update the icon
win32gui.UpdateResource(
    win32api.BeginUpdateResource(exe_path, False),
    win32con.RT_ICON,
    1,  # Icon ID
    open(icon_path, "rb").read()
)
win32api.EndUpdateResource(win32api.BeginUpdateResource(exe_path, False), False)
```
