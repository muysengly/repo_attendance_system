```python
import sqlite3
```

```python
conn = None
try:
    conn = sqlite3.connect("../database.sqlite")
    print("Connected")
except sqlite3.Error as e:
    print(e)
```

```python
# create a table
conn.execute(
    """
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        age INTEGER NOT NULL
    );
    """
)
conn.commit()
print("Table created")
```

```python
# read data
cursor = conn.execute("SELECT * FROM users")
for row in cursor.fetchall():
    print(row)
```

```python
# add data
conn.execute(
    """
    INSERT INTO users (name, age)
    VALUES 
        ('Alice', 30),
        ('Bob', 25),
        ('Charlie', 35)
    """
)
conn.commit()
print("Data inserted")

# commit the changes


# read data
cursor = conn.execute("SELECT * FROM users")
for row in cursor.fetchall():
    print(row)
```

```python
# delete data
conn.execute(
    """
    DELETE FROM users 
    WHERE id = 3
    """
)
conn.commit()
print("Data deleted")

# read data again
cursor = conn.execute("SELECT * FROM users")
for row in cursor.fetchall():
    print(row)
```

```python
# update data
conn.execute(
    """
    UPDATE users 
    SET age = 31 
    WHERE name = 'Alice'
    """
)
conn.commit()
print("Data updated")

cursor = conn.execute("SELECT * FROM users")
for row in cursor.fetchall():
    print(row)
```

```python
# close the connection
conn.close()
```
