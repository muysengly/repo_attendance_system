```python
class Window(Ui_MainWindow, QMainWindow):
    def __init__(self):
        super().__init__()
        self.setupUi(self)
        self.setWindowFlags(self.windowFlags() | Qt.WindowStaysOnTopHint)
        self.setWindowIcon(QIcon(f"{path_depth}resource/asset/itc_logo.png"))
        self.setWindowTitle("Add New Group Form")


        self.show()

```

```python
app = QApplication([])

win = Window()


def click_add():

    group_name = win.lineEdit_group_name.text()

    if group_name.__len__() >= 6:
        if not os.path.exists(path_depth + "resource/database/" + group_name + ".pkl"):
            pickle.dump([], open(path_depth + "resource/database/"+ group_name + ".pkl", "wb"))
            win.label_status.setText(f"Group '{group_name}' created successfully!")
        else:
            win.label_status.setText(f"Group '{group_name}' already exists!")
    else:
        win.label_status.setText(f"At least 6 characters!")


win.pushButton_add.clicked.connect(click_add)

app.exec_()


app = None
```
