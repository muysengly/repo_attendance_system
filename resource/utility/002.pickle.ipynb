```python
import cv2
import pickle
import matplotlib.pyplot as plt
```

```python
# read image from file

img1 = cv2.imread("img1.png")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

plt.imshow(img1)
plt.axis("off")
plt.show()
```

```python
# read image from file

img2 = cv2.imread("img2.png")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.imshow(img2)
plt.axis("off")
plt.show()
```

```python
# combine images into a list

images = [img1, img2]
```

```python
# save images to a pickle file

pickle.dump(images, open("images.pkl", "wb"))
```

```python
# load images from the pickle file

images = pickle.load(open("images.pkl", "rb"))
```

```python
plt.imshow(images[0])
plt.axis("off")
plt.show()
```
