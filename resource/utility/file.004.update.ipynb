```python
# clone github as zip file
import requests


url = f"https://github.com/muysengly/repo_attendance_system/archive/refs/heads/main.zip"
output_file = f"files.zip"

response = requests.get(url)
if response.status_code == 200:
    with open(output_file, "wb") as f:
        f.write(response.content)
    print(f"Downloaded {output_file}")
else:
    print(f"Failed to download: {response.status_code}")
```

```python
# extract the files.zip
import zipfile

with zipfile.ZipFile("files.zip", "r") as zip_ref:
    zip_ref.extractall("__tmp__")
```

```python
# extract a specific folder
import zipfile

def extract_folder(zip_file, folder_name, output_dir):
    with zipfile.ZipFile(zip_file, "r") as zip_ref:
        for member in zip_ref.namelist():
            if member.startswith(folder_name + "/"):
                zip_ref.extract(member, output_dir)

zip_file = "files.zip"
folder_name = "repo_attendance_system-main/resource"
output_dir = "abc"
extract_folder(zip_file, folder_name, output_dir)
```
