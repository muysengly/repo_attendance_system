```python
import cv2


img_name = r"SAM SINATH\data_2025_04_03_13_12_22.jpg"

# Load an image (ACDSee export or any other source)
image = cv2.imread(img_name)


# Convert from RGB to BGR (if needed, OpenCV loads images in BGR by default)
bgr_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Save the BGR image
cv2.imwrite(img_name, bgr_image)
```
